```python
#Tabelle (Die Tabellenelemente) ausgeben lassen
import requests
from bs4 import BeautifulSoup

# URL der Zielseite
url = "https://www.8notes.com/viola/rock_and_pop/"

# HTML-Seite abrufen
response = requests.get(url)

# Prüfen, ob der Abruf erfolgreich war
if response.status_code == 200:
    # HTML-Inhalt parsen
    soup = BeautifulSoup(response.text, 'html.parser')

    # Alle <tr> Tags mit onclick Attribut suchen
    rows = soup.find_all('tr', onclick=True)

    # Zeilen ausgeben, die mit <tr onclick="document.location= beginnen
    for row in rows:
        if row['onclick'].startswith("document.location="):
            print(row)
else:
    print(f"Fehler beim Abrufen der Seite: {response.status_code}")
```

```python
from bs4 import BeautifulSoup

# Das gegebene HTML als String # Intendationen dürfen so bleiben !!! #Kopierfüge den tbody einfach ein!!!
html = '''
<table>
<tr onclick="document.location='/scores/12124.asp'">
    <td class="piece_type ditchit"><div><img src="/im/8notes-circle-purple-sm.svg" class="scoreicon" title="David Bruce Cool Blues Sheet Music"></div></td>
    <td class="artname">David Bruce</td>
    <td class="fsmtitle">Cool Blues</td>
    <td class="level_type ditchit"><img title="Beginners Level" alt="Beginners Level" src="/images/eg1c2.gif" class="difflevel"></td>
</tr>
<tr onclick="document.location='/scores/12125.asp'">
    <td class="piece_type ditchit"><div><img src="/im/8notes-circle-purple-sm.svg" class="scoreicon" title="Mozart Symphony No. 40"></div></td>
    <td class="artname">Mozart</td>
    <td class="fsmtitle">Symphony No. 40</td>
    <td class="level_type ditchit"><img title="Intermediate Level" alt="Intermediate Level" src="/images/eg1c3.gif" class="difflevel"></td>
</tr>
<tr onclick="document.location='/scores/12126.asp'">
    <td class="piece_type ditchit"><div><img src="/im/8notes-circle-purple-sm.svg" class="scoreicon" title="Beethoven Moonlight Sonata"></div></td>
    <td class="artname">Beethoven</td>
    <td class="fsmtitle">Moonlight Sonata</td>
    <td class="level_type ditchit"><img title="Advanced Level" alt="Advanced Level" src="/images/eg1c4.gif" class="difflevel"></td>
</tr>
</table>
'''

# BeautifulSoup-Objekt erstellen
soup = BeautifulSoup(html, 'html.parser')

# Alle tr-Elemente in der Tabelle finden
entries = soup.find_all('tr')

# Den 'title' von class="piece_type ditchit" extrahieren
title = soup.find(class_="piece_type ditchit").find('img')['title']
print("Title:", title)

# Den 'Schwierigkeitsgrad' von class="level_type ditchit" extrahieren
schwierigkeitsgrad = soup.find(class_="level_type ditchit").find('img')['alt']
print("Schwierigkeitsgrad:", schwierigkeitsgrad)

print(f'{schwierigkeitsgrad}, {title}')
print("-----------Nummern-Schreibweise:-------------")
# Schleife über alle Einträge, um eine Nummerierte Ausgabe aller zu bekommen
for index, entry in enumerate(entries, start=1):
    # Den 'title' von class="piece_type ditchit" extrahieren
    piece_type_title = entry.find(class_="piece_type ditchit").find('img')['title']
    # Formatierung der Nummerierung mit führender Null für einstellige Zahlen
    entry_number = f"{index:02}"
    print(f"{entry_number}: {piece_type_title}")

#Die Links können mit Regex extrahiert werden siehe Chatgpt # Wir wollen aber rein-Python based.
```
